```python
import tensorflow as tf
import numpy as np
import pickle
```

```python
model = tf.keras.models.load_model('ProfitPredictor.h5')
```

```python
scFeatures = pickle.load(open('scFeatures','rb'))
minMaxLabel = pickle.load(open('minMaxLabel','rb'))
```

```python
ohe = pickle.load(open("StateEncoderDecoder.pkl","rb"))

# Deploy Code

rdSpend = float(input("Enter RD Spend: "))
admSpend = float(input("Enter Admin Spend: "))
markSpend = float(input("Enter Marketing Spend: "))
state = input("Enter State: ")

if state in ohe.categories_[0]:
  encodedState = ohe.transform(np.array([[state]]))

  # 2. Represent the input in the form of 2d np array
  finalFeature = np.concatenate( (np.array([[rdSpend,admSpend,markSpend]]), encodedState) , axis = 1)
  print(finalFeature)
  print(type(finalFeature))

  # 3. Apply Standardization
  finalFeature = scFeatures.transform(finalFeature)
  print(finalFeature)
  print(type(finalFeature))

  # 4. Get the Prediction
  predictProfit = model.predict(finalFeature)

  # 5. Inverse transform predicted label
  predictProfit = minMaxLabel.inverse_transform(predictProfit)

  # 6.Output

  print("Predicted profit is $ {}".format(predictProfit))

else:
  print("Given state {} is not supported by the model ".format(state))
```
